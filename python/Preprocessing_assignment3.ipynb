```python
import geopandas as gpd
import pandas as pd 
#pd.set_option("display.max_rows", None)
```

```python
circo_df = gpd.read_file("data/circoscrizioni.json")
circo_df = circo_df[['numero_cir', 'nome', 'area', 'geometry']]
circo_df.head()
```

```python
circo_geometry_col = circo_df["geometry"]
circo_geometry_col
```

```python
circo_area_col = circo_df["area"]
circo_area_col
```

```python
tree_df = gpd.read_file("data/geo_data_trees.geo.json")
tree_df.head()
```

```python
tree_df = tree_df[['Tree ID', 'Canopy Cover (m2)', 'Oxygen Production (kg/yr)', 'geometry']]
tree_df.head()
```

```python
def get_circoscrizione(tree_point):
    for idx_polygon, polygon in circo_geometry_col.items():
        if tree_point == None: return "None"
        if tree_point.within(polygon):
            return  circo_df['nome'][idx_polygon]
```

```python
def get_area():
    for idx_polygon, polygon in circo_geometry_col.items():
        if tree_point == None: return "None"
        if tree_point.within(polygon):
            return  circo_df['nome'][idx_polygon]
```

```python
tree_df2 = tree_df.copy()
tree_df2['circoscrizione'] = ""
tree_df2['circoscrizione']= tree_df2['geometry'].map(get_circoscrizione)

tree_df2.head()
```

```python
tree_df2['area'] = ""
tree_df2['area']= tree_df2['circoscrizione'].map(get_area)

tree_df2.head()
```

```python
tree_df4 = tree_df2.copy()
tree_df4['count'] = tree_df4.groupby(['circoscrizione'])['circoscrizione'].transform('count')
tree_df4['density'] = tree_df4.groupby(['circoscrizione'])['Canopy Cover (m2)'].transform('sum')

tree_df4 = tree_df4.drop(["geometry", "area"], axis=1)
tree_df4 = tree_df4.drop_duplicates()
tree_df4 = tree_df4.sort_values(by='circoscrizione', ascending=True)

tree_df4
```
