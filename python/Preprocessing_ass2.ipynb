```python
import geopandas as gpd
import pandas as pd 
pd.set_option("display.max_rows", None)
```

```python
circo_df = gpd.read_file("data/circoscrizioni.json")
circo_df.head()
#circo_df = circo_df[['numero_cir', 'nome', 'geometry']]
#circo_df.head()
```

```python
df = gpd.read_file("data/geo_data_trees.geo.json")
tree_df = df[["Name", "Height (m)", "Canopy Cover (m2)", "Oxygen Production (kg/yr)", "geometry"]]
tree_df.head()
```

```python
circo_geometry_col = circo_df["geometry"]

def get_circoscrizione(tree_point):
    for idx_polygon, polygon in circo_geometry_col.items():
        if tree_point == None: return "None"
        if tree_point.within(polygon):
            return  circo_df['nome'][idx_polygon]
        

tree_df2 = tree_df.copy()
tree_df2['circoscrizione'] = ""
tree_df2['circoscrizione']= tree_df2['geometry'].map(get_circoscrizione)
tree_df2 = tree_df2.drop(columns=['geometry'])

tree_df2.head()
```

```python
occurrences = dict()
for i in range(df.shape[0]):
    name = df.iloc[i][1]
    if name in occurrences:
        occurrences[name]=occurrences[name]+1
    else:
        occurrences[name]=1
        
trees_by_occurrences = sorted(occurrences.items(), key=lambda x:x[1], reverse=True)

top_six = []
for i in range(6):
    top_six.append(trees_by_occurrences[i][0])
    
print(top_six)    
```

```python
ass2_df = tree_df2[tree_df2['Name'].isin(top_six)]
#ass2_df = tree_df2.where(tree_df2['Name'] in top_six)
ass2_df = ass2_df.sort_values(['Name'])
ass2_df.head(500)
```

```python
ass2_df = ass2_df.rename(columns={"Height (m)": "Height", "Canopy Cover (m2)": "Canopy", "Oxygen Production (kg/yr)": "CO2"})
ass2_df.head()
```
