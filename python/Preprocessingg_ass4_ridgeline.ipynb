```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("assignment_4_salorno/salorno.csv")
```

```python
df.head()
```

```python
df = df[['day', 'min_temp', 'max_temp']]
df["avg_temp"] = np.around((df["min_temp"] + df["max_temp"] )/2, decimals=2)
df.head()
```

```python
df.index = pd.to_datetime(df['day'], format='%d.%m.%Y')
```

```python
df.head()
```

ogni colonna:
    un mese gennaio - dicembre
ogni riga:
    temperature di ogni gg (30 valori)

```python
years = [1993, 1997, 2001, 2005, 2009, 2013, 2017, 2021]
data_min_dict = {'year':[]}
data_max_dict = {'year':[]}
for y in years:
    counter = 31
    while counter != 0:
        data_min_dict['year'].append(y)
        data_max_dict['year'].append(y)
        counter -= 1

for x in range(1,13):
    data_min_dict[x] = []
    data_max_dict[x] = []
```

```python
for row in df.iterrows():
    date = row[0]
    if date.year not in years:
        continue
    tmin = row[1][1]
    tmax = row[1][2]
    data_min_dict[date.month].append(tmin)
    data_max_dict[date.month].append(tmax)
```

```python
# filling with nan all the missing values
for m in range(2, 13):
    len_d = len(data_min_dict[m])
    while len_d != 248:
        data_min_dict[m].append('None')
        len_d = len(data_min_dict[m])

for m in range(2, 13):
    len_d = len(data_max_dict[m])
    while len_d != 248:
        data_max_dict[m].append('None')
        len_d = len(data_max_dict[m])
```

```python
df_min = pd.DataFrame(data_min_dict)
df_max = pd.DataFrame(data_max_dict)
df_min.rename(columns={1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sept', 10:'Oct', 11:'Nov', 12:'Dec'}, inplace=True)
df_max.rename(columns={1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sept', 10:'Oct', 11:'Nov', 12:'Dec'}, inplace=True)
```
