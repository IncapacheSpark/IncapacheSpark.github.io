```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("salorno.csv")
df = df.astype({"min_temp":"float", "max_temp":"float" })
df = df[["day", "min_temp", "max_temp"]]
```

```python
df["avg_temp"] = np.around((df["min_temp"] + df["max_temp"] )/2, decimals=2)
df['year'] = df['day'].str[-4:]
df['month_year'] = df['day'].str[3:10]

years_to_save = ["1993", "1997", "2001", "2005", "2009", "2013", "2017", "2021"]
df = df.apply(lambda row: row[df['year'].isin(years_to_save)])

df
```

```python
df['sum_temp_month'] = df.groupby(['month_year'])['avg_temp'].transform('sum')
df['length_of_month'] = df.groupby(['month_year'])['month_year'].transform('count')
df['monthly_temp'] = df['sum_temp_month'] / df['length_of_month']
df = df[['monthly_temp', 'month_year'
df.to_csv('line_chart_dataset.csv')
df
```
