```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/IncapacheSpark.github.io/python/assignment_4_salorno/salorno.csv")
df = df.astype({"min_temp":"float", "max_temp":"float" })
df = df[["day", "min_temp", "max_temp"]]
```

```python
df
```

```python
df["avg_temp"] = np.around((df["min_temp"] + df["max_temp"] )/2, decimals=2)
```

```python
df.index = pd.to_datetime(df['day'], format='%d.%m.%Y')
# df.groupby(pd.Grouper(freq='M'))
```

```python
df = df[["min_temp", "max_temp", "avg_temp"]]
df
```

```python
years = [1993, 1997, 2001, 2005, 2009, 2013, 2017, 2021]
filtered_df = df[df.index.year.isin(years)]
```

```python
filtered_df
```

```python
filtered_df.index = filtered_df.index.to_period('M')
```

```python
filtered_df_2 = filtered_df.groupby(filtered_df.index).min()
filtered_df_2
```

```python
filtered_df_2['max_temp'] = filtered_df.groupby(filtered_df.index).max()['max_temp']
filtered_df_2['avg_temp'] = np.around(filtered_df.groupby(filtered_df.index).mean()['avg_temp'], decimals=2)
filtered_df_2
```

```python
filtered_df_2.to_csv('/content/IncapacheSpark.github.io/python/data/assignment4.csv')
```
