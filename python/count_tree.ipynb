```python
occurrences = dict()
for i in range(df.shape[0]):
    name = df.iloc[i][1]
    if name in occurrences:
        occurrences[name]=occurrences[name]+1
    else:
        occurrences[name]=1


trees_by_occurrences = sorted(occurrences.items(), key=lambda x:x[1], reverse=True)
```

```python
for elem in trees_by_occurrences:
    print(elem)
```
