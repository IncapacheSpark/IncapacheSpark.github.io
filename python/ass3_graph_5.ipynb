```python
import geopandas as gpd
import pandas as pd
#pd.set_option("display.max_rows", None)
```

```python
tree_df = gpd.read_file("data/geo_data_trees.geo.json")
tree_df.head()
```

```python
tree_df[['Tree ID', 'Name']]
```

```python
tree_count = gpd.GeoDataFrame(tree_df['Name']) #, 'Latitude', 'Longitude']])
tree_count['count'] = tree_count.groupby('Name')['Name'].transform('count')
# tree_top_ten.drop(['DBH (cm)', 'Repl'], axis=1, inplace=True)
tree_count.drop_duplicates(subset=["Name", "count"], inplace=True)
tree_count.sort_values(by=['count'], inplace=True, ascending=False)
tree_count.head()
```
